```python
import numpy as np

# Calcular a matriz de correlação
correlation_matrix = df.corr()

# Exibir a matriz de correlação
print(correlation_matrix)

# Visualizar a matriz de correlação usando um mapa de calor
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Matriz de Correlação')
plt.show()
```

```python
#analisando
print("1 indica uma correlação positiva perfeita (quando uma variável aumenta, a outra também aumenta na mesma proporção).")
```

```python
print("0 indica ausência de correlação linear.")
```

```python
print("-1 indica uma correlação negativa perfeita (quando uma variável aumenta, a outra diminui na mesma proporção).")
```

```python
print("- 0,55 sugere que há uma tendência para uma variável aumentar enquanto a outra diminui, mas essa relação não é perfeita, pois o valor não atinge -1. No caso específico de -0,55, a correlação negativa moderada sugere que à medida que uma variável aumenta, a outra tende a diminuir, mas a relação não é tão forte.")
```

```python
print("Portanto, um valor de -0,88 sugere que há uma relação muito forte e negativa entre as variáveis, onde o aumento em uma está fortemente associado à diminuição na outra.")
```
